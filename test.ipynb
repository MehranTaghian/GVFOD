```python
path = 'exp/exp_gvfod/csv/train_size_delay_0_default_False_trainloss_False_loose_l1.csv'
```

```python
res = pd.read_csv(path)
res
```

```python
algorithms = pd.unique(res['Algorithm']) 

fig, ax = plt.subplots(2, 6, figsize=[16, 5])
for i, alg in enumerate(algorithms):
    row, col = int(i / 6), int(i % 6)
    sns.lineplot(
                x='Training Size',
                y=f"f1_score_loose_l1",
                ci=95,
                data=res[res['Algorithm'] == alg],
                ax=ax[row, col],
                legend="brief",
                label=alg
            )
    ax[row, col].set_ylim(0, 1)
    if col > 0:
        ax[row, col].set_ylabel('')
    if row == 0:
        ax[row, col].set_xlabel('')
        
        
ax[-1, -1].axis('off')
        
```

```python
sns.set_theme()
sns.set_palette("cividis")
fig, ax = plt.subplots(figsize=[12, 10])


sns.lineplot(
            x='Training Size',
            y=f"f1_score_loose_l1",
            hue='Algorithm',
            style='Class',
            ci=95,
            data=res,
            ax=ax,
            legend="brief"
            )
```

```python
metric = 'f1_score_loose_l1'
var = res[['Algorithm', 'Training Size', metric]]
var = var.groupby(['Algorithm', 'Training Size']).agg({metric: 'var'}).reset_index()
```

```python
algorithms, training_size = pd.unique(var['Algorithm']), pd.unique(var['Training Size'])

var[metric] = (var[metric] - var[metric].min()) / (var[metric].max() - var[metric].min())

for a in algorithms:
    for t in training_size:
        variance = var.loc[(var['Algorithm'] == a) & (var['Training Size'] == t)][metric].iloc[0]
        for _ in range(int(100 * variance)):
            new_row = {'Algorithm': a, 'Training Size': t, metric: 0}
            var = var.append(new_row, ignore_index=True)
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
sns.violinplot(data=var, x="Algorithm", y="Training Size", scale="count", cut=0, ax=ax)
ax.set_xlabel('Algorithm', fontdict=font1)
```
