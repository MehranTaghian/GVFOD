```python
path = 'exp/exp_gvfod/figures/res.csv'
```

```python
res = pd.read_csv(path)
res
```

```python
algorithms = pd.unique(res['Algorithm']) 

fig, ax = plt.subplots(2, 6, figsize=[16, 5])
for i, alg in enumerate(algorithms):
    row, col = int(i / 6), int(i % 6)
    sns.lineplot(
                x='Training Size',
                y=f"f1_score_loose_l1",
                ci=95,
                data=res[res['Algorithm'] == alg],
                ax=ax[row, col],
                legend="brief",
                label=alg
            )
    ax[row, col].set_ylim(0, 1)
    if col > 0:
        ax[row, col].set_ylabel('')
    if row == 0:
        ax[row, col].set_xlabel('')
        
        
ax[-1, -1].axis('off')
        
```

```python
sns.set_theme()
sns.set_palette("cividis")
fig, ax = plt.subplots(figsize=[12, 10])


sns.lineplot(
            x='Training Size',
            y=f"f1_score_loose_l1",
            hue='Algorithm',
            style='Class',
            ci=95,
            data=res,
            ax=ax,
            legend="brief"
            )
```
